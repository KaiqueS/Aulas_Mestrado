# Gary King - Replication, Replication

<b>Problems In Empirical Political Science</b>

King argues that it is only possible to fully understand an emp. analysis when the process used on the latter is completely accessible to any other researcher. By process, the author means both data generation and data analysis. To be precise, it is not required that a replication to be indeed performed, but only for replications to be feasible. 

The author says that it is not rare for works which do not follow the replication standard to face problems. Following the standard, even if no intention to provide replicability, may lead the reasearch to discover problems in its own analysis or data collection process. Aside from that, not following the standard itself harms research progress. If results cannot be guaranteed to be true, what is the purpose of trying to read a work with those results? 

Replication by itself might lead to better research comprehension and evaluation, because how one gets and measure data by itself provides information about the results, and, consequently, about the world to which the latter are related. Proceeding as proposed is the best way to further advance with science, since, by allowing for replication, it is possible to build new methodologies upon current data or even to add more data to an existent dataset. Replication cuts the cost of advancing with research by making the latter easier, whereas no replicability makes it more expensive for an study to be used, since no one will waste its time with something that might not be right after all.

<b>Proposed Solutions</b>

King proposes solutions applicable by the many parts that are involved, directly or not, in the scientific endeavor.

1. <b>Authors</b>

Authors should follow the replication standard and provide as much information as possible about how they proceeded to gather and analyze data. To do so, the authors must:

* Create a Replication Data Set, which must include all data necessary to replicate the results of the research, and provide public access to it. King proceeds babbling about how and where to store data digitally and whatsoever
    

2. <b>Tenure And Promotion Review Committees</b>

Its purpose is to promote chosen candidates. The choice must be made based on wether and how much a candidate adheres to the replication standard

3. <b>Graduate Programs</b>

Graduation programs should induce its students to adhere to the replication standard, either in its courses( e.g., by making students replicate published works' results ), or by proposing rules, to be followed by any student when writing their dissertation, that are in accord with the replication standard.

4. <b>Editors And Reviewers Of Books And Journals</b>

Those must try to enforce replicability by requiring submitted work to adhere to the replication standard. This can be done, e.g., by requiring specification of where the author will provide its data and when. Also, journals could also require more reviewers to try and replicate submitted but unpublished papers, so as to make it possible to give some feedback to authors, where the latter might, then, make further ajudstments to their papers.

<b>Important Exceptions</b>

Exceptions to full submition of data.

 1. Confidentiality: There are some cases, beyond the mere avoidance of publishing names and addresses or any personal information of respondents, where data must not be provided. Whenever any extra sensible information is present, or when data is proprietary, confidentiality must be respected, and the corresponding data must not be publicly provided. 
 
 
 2. Right Of First Publication: Sometimes the data for a research might not be imediately provided, so as to grant the right of first publication, i.e., that the person that collected the data must be the first to use it in a publication. When this happens, the research might ask for an embargo for a period of time before the data is made public. Those cases must be rare, and, it might be made necessary to submit data exclusively used on the research, so as to make it possible for the results to be verified and replicated. Sometimes, this embargo might also be used so as to make it possible for the researcher to better work and format a subset of the data intended to be used in its study. In the latter cases, a copy of the old not-formatted data must also be submitted.
 
<b>Support For These Policies</b>

Those recommendations are growing in adoption, be it by journals, graduate programs, and whatsoever.

<b>2 - Replicar É Preciso</b>

Há três dimensões na replicabilidade, quais sejam:

* Substantiva: aprimoramento da qualidade nas pesquisas e proporção do acúmulo de conhecimento

* Pedagógica: viabilizar o estudo, por parte de pós-graduandos, de análise de dados e suas técnicas

* Transparência: expor erros honestos ou desonestos, de forma que seja possível se proteger dos mesmos

<b>2.1 O Que É Replicabilidade</b>

Para os autores, de acordo com King, replicabilidade é o fornecimento, por parte do pesquisador, de informações suficientes para se entender, avaliar e replicar resultados de um estudo. Essas informações são justamente um passo-a-passo sobre como foram coletados e analisados os dados utilizados na pesquisa.

<b>NdR 12:</b> replicação, reanálise, verificação e análise secundária são conceitos diferentes. 

* Replicação: coleta independente de dados

* Reanálise: mesmo estudo, não necessariamente mesma base de dados

* Verificação: mesmo estudo, mesma base de dados

* Análise secundária: mesmo banco de dados, problema diferente

<b>2.2 Dimensão Substantiva</b>

Existem algumas vantagens de se disponibilizar um estudo replicável. A primeira, segundo GMS( 2003 ), é que autores que fazem trabalhos replicáveis são mais citados. A segunda é a própria replicabilidade e uma consequência didática desta, qual seja, a possibilidade de se desenvolver melhores análises e da exposição de possíveis erros em estudos já feitos. A maior possibilidade de expor erros garante uma maior falseabiliade dos estudos feitos, o que seria outra vantagem da replicabilidade.

Os autores também afirmam que Meier( 1995 ) identifica outros dois pontos, quais sejam, um aumento na validade e credibilidade do trabalho feito.

<b>2.3 Dimensão Pedagógica</b>

Replicabilidade tem um grande efeito didático, isto se dá pois, quando a pesquisa é replicável, estudantes podem refazer os passos que levaram o autor ao resultado obtido na própria pesquisa.

<b>2.4 Dimensão Transparência</b>

Segundo os autores, Charles Babbage identifica três modos de fraude:

1. Aparamento: ajuste, ou manipulação, de dados

2. Peneiramento: seleção apenas dos resultados que confirmam a hipótese

3. Falsificação: invenção de dados, com a intenção de descrever um padrão

<b>O Caso Do SIGOBr</b>

Tem como objetivo integrar práticas desejáveis de replicabilidade em um único produto. Para isso, compõe-se por cinco atividades principais:

* Compilação de base de dados: junção de dados de diferentes bancos de dados e diferentes temas, em níveis estaduais e municipais

* Replicabilidade: incentivar replicação de estudos publicados. Tem como objetivo aprimorar pesquisas, educar alunos e verificar resultados já estabelecidos.

* Repositório de dados: servir como repositório de dados

* Cursos de capacitação: fornecer treinamento técnico

* Divulgação: exposição de resultados com o intuito de tornar os mesmos públicos.
